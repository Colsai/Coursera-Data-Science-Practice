```python
#Retrieve Website for Data
data= 'https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/coursera/Topic_Survey_Assignment.csv'

#Read CSV as dataframe 'dss_df'
dss_df = pd.read_csv(data, index_col=0) #the new dataframe is called 'dss_df'

#Set and round to two decimal places 
dss_df_pct = round(dss_df/2233, 2)

#Sort the values of 'Very Interested', Descending
dss_df_pct.sort_values(['Very interested'], ascending=False, inplace=True)

#Label the Graph
ax = dss_df_pct.plot(kind='bar', width=.8, figsize = (40,8), color=['#5cb85c', '#5bc0de', '#d9534f'])
ax.set_title('Percent of People Interested in Data Science')
ax.set_xlabel('Program Type')
ax.set_ylabel('Percent')
#ax.set_rc('font', size=16)

plt.annotate('',                      # s: str. will leave it blank for no text
             xy=(32, 70),             # place head of the arrow at point (year 2012 , pop 70)
             xytext=(28, 20),         # place base of the arrow at point (year 2008 , pop 20)
             xycoords='data',         # will use the coordinate system of the object being annotated 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

plt.show()
```

```python
#Retrieve Website for Data
crime_data= 'https://cocl.us/sanfran_crime_dataset'

#Set as Dataframe 'crime_df'
crime_df = pd.read_csv(crime_data)

#This code sorts by Value_counts the counts of each district, then orders it in descending order
crime_area_df = pd.value_counts(crime_df.PdDistrict).to_frame().reset_index() #as new df: crime_area_df
crime_area_df.sort_values(['PdDistrict'], ascending = False, inplace= True)

#Change the Column Names to clean up the data for use
crime_area_df.rename(columns = {'index':'Neighborhood', 'PdDistrict':'Count'}, inplace = True)

crime_area_df
```

```python
# download countries geojson file
!wget --quiet https://cocl.us/sanfran_geojson -O san-francisco.geojson
world_geo = r'san-francisco.geojson' # geojson file

# San Francisco latitude and longitude values
latitude = 37.77
longitude = -122.42

sanfran_map = folium.Map(location=[latitude, longitude], zoom_start=12)

# display the map of San Francisco
sanfran_map.choropleth(
    geo_data=world_geo, #the geoJson file
    data=crime_area_df, #the dataframe from before
    columns=['Neighborhood', 'Count'],
    key_on='feature.properties.DISTRICT', #From the geoJson file
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='San Francisco Crime, by District'
)

#finally, display the map
sanfran_map
```
