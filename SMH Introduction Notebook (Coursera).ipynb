<h3/> The function of my code is to tell the current time for when someone visits this introductory notebook. <br> Since I am new to Python, I found the code on https://www.w3resource.com/python-exercises/python-basic-exercise-3.php.

```python
#imports the time
import datetime

#converts to variable 'now'
now = datetime.datetime.now()

#prints as sentence with variables
print ("You have visited at " + now.strftime("%H:%M") + "(UTC) on " + now.strftime("%m/%d/%y") + ".")
```
